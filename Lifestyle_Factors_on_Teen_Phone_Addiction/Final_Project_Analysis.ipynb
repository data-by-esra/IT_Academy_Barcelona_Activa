```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = df.drop(columns=["ID"])
corr = df1.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="viridis", linewidths=0.8)
plt.xticks(fontsize=11, fontweight='bold', color='darkblue')
plt.yticks(fontsize=11, fontweight='bold', color='darkblue')
plt.tight_layout()
plt.show()
```

```python

df2 = df.drop(columns=["ID"])
corr = df2.corr(numeric_only=True)

corr_with_addiction = corr.drop(index='Addiction_Level') 
corr_with_addiction = corr_with_addiction[['Addiction_Level']]  

plt.figure(figsize=(12, 10))
sns.heatmap(corr_with_addiction.sort_values('Addiction_Level'), 
            annot=True, fmt=".2f", cmap="seismic", linewidths=0.8, vmin=-1, vmax=1)
plt.xticks(fontsize=11, fontweight='bold', color='darkblue')
plt.yticks(fontsize=11, fontweight='bold', color='darkblue')
#plt.title("Addiction Level vs Other Numeric Variables", fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

Addiction Level

```python
sns.boxplot(x='Age',y='Addiction_Level',data=df,palette='viridis')
plt.xticks(fontsize=11, fontweight='bold', color='darkblue')
plt.yticks(fontsize=11, fontweight='bold', color='darkblue')
plt.xlabel('Age',fontsize=12,fontweight='bold',color='darkblue')
plt.ylabel('Addiction Level',fontsize=12,fontweight='bold', color='darkblue')
plt.tight_layout()
plt.grid(True)
plt.show()
```

Daily Phone Usage 

```python
import numpy as np
daily_usage_hours = df['Daily_Usage_Hours']


counts, bins = np.histogram(daily_usage_hours, bins=20)


cmap = plt.cm.Blues  
norm = plt.Normalize(counts.min(), counts.max())

fig, ax = plt.subplots(figsize=(8, 5))

for i in range(len(bins) - 1):
    ax.bar(
        bins[i],
        counts[i],
        width=bins[i+1] - bins[i],
        color=cmap(norm(counts[i])),
        edgecolor="black"
    )

ax.set_xlabel("Daily Usage Hours",fontweight='bold', color='darkblue')
ax.set_ylabel("",fontweight='bold', color='darkblue')
plt.xticks(fontsize=12, fontweight='bold', color='darkblue')
plt.yticks(fontsize=12, fontweight='bold', color='darkblue')

plt.show()
```

```python
plt.figure(figsize=(7, 5))
colors = {'Male': "#0B8BD5", 'Female': "#e73cd3", 'Other': "#3ed16a"}
sns.scatterplot(data=df, 
                x="Daily_Usage_Hours", 
                y="Addiction_Level", hue="Gender",
                alpha=0.7, 
                palette= colors, 
                edgecolor="black")

plt.xticks(fontsize=12, fontweight='bold', color='darkblue')
plt.yticks(fontsize=12, fontweight='bold', color='darkblue')
plt.xlabel("Daily Usage Hours", fontsize=10,fontweight='bold', color='darkblue')
plt.ylabel("Addiction Level", fontsize=10,fontweight='bold', color='darkblue')
plt.tight_layout()
plt.show()
```

Summary

-Addiction tends to increase as usage increases â†’ strong positive trend.

-There is a notable cluster at the highest addiction level (10).

-Gender does not show a significant effect on addiction levels.

-Some individuals show high addiction even with low daily usage.

SLEEP HOURS

```python
import matplotlib.pyplot as plt
import numpy as np


sleep_hours = df['Sleep_Hours']


counts, bins = np.histogram(sleep_hours, bins=20)


cmap = plt.cm.Purples 
norm = plt.Normalize(counts.min(), counts.max())

fig, ax = plt.subplots(figsize=(8, 5))

for i in range(len(bins) - 1):
    ax.bar(
        bins[i],
        counts[i],
        width=bins[i+1] - bins[i],
        color=cmap(norm(counts[i])),
        edgecolor="black"
    )

ax.set_xlabel("Sleep Hours",fontweight='bold', color='darkblue')
ax.set_ylabel("",fontweight='bold', color='darkblue')
plt.xticks(fontsize=12, fontweight='bold', color='darkblue')
plt.yticks(fontsize=12, fontweight='bold', color='darkblue')

plt.show()
```

```python
plt.figure(figsize=(7, 5))
colors = {'Male': "#0B8BD5", 'Female': "#e73cd3", 'Other': "#3ed16a"}
sns.scatterplot(data=df, 
                x="Sleep_Hours", 
                y="Addiction_Level", hue="Gender",
                alpha=0.7,  
                edgecolor="black",palette= colors)

plt.xlabel("Sleep Hours", fontsize=12,fontweight='bold', color='darkblue')
plt.ylabel("Addiction Level", fontsize=12,fontweight='bold', color='darkblue')
plt.xticks(fontsize=12, fontweight='bold', color='darkblue')
plt.yticks(fontsize=12, fontweight='bold', color='darkblue')

plt.tight_layout()
plt.show()
```
