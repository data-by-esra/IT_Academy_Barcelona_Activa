Exercise 6 : 1.Create a new column that is the date of birth in Datetime format from the columns day, month and year. Using this column create a function that, given a date, calculates your current age as of today.
2.Use the function you just created to generate a new column in the DataFrame with the current age.


```python
import pandas as pd
from datetime import datetime

df["Date de Naixament"] = pd.to_datetime( df.rename(columns = {'Any':'year','Mes':'month','Dia':'day'})[["year","month","day"]])

def calculate_age(birth_date,today = datetime.today()):
    age = today.year - birth_date.year
    # If birthday hasn't occurred yet this year, subtract 1
    if (today.month, today.day) < (birth_date.month, birth_date.day):
        age -= 1
    return age

df["Edat Actual"] = df["Date de Naixament"].apply(calculate_age)
df
```

Exercise 1:1.Create a function that takes a dataframe as an input parameter.
The function must create (and export) a graph automatically for each column of the dataframe. For example:
   -a histogram/boxplot if the variable is numeric
   -a few bars of the most frequent values ​​if it is categorical
   -a few bars of the most frequent years if the data is in date format.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

def auto_graphs(df, output_folder="graphs"):
    os.makedirs(output_folder, exist_ok=True)
    for column in df.columns:
        plt.figure(figsize=(6, 4))
        
        # Drop missing values for the column
        data = df[column].dropna()
        
        # Detect column type and plot accordingly
        if pd.api.types.is_numeric_dtype(data):
            # Numeric variable → histogram and boxplot
            plt.subplot(2, 1, 1)
            sns.histplot(data, kde=True)
            plt.title(f"{column} - Histogram")

            plt.subplot(2, 1, 2)
            sns.boxplot(x=data)
            plt.title(f"{column} - Boxplot")

        elif pd.api.types.is_datetime64_any_dtype(data):
            # Date variable → most frequent years
            years = data.dt.year.value_counts().nlargest(10)
            sns.barplot(x=years.index, y=years.values)
            plt.title(f"{column} - Most Frequent Years")
            plt.xlabel("Year")
            plt.ylabel("Frequency")

        else:
            # Categorical variable → most frequent values
            top_values = data.value_counts().nlargest(10)
            sns.barplot(x=top_values.values, y=top_values.index)
            plt.title(f"{column} - Top 10 Most Frequent Values")
            plt.xlabel("Count")
            plt.ylabel(column)

        plt.tight_layout()
        save_path = os.path.join(output_folder, f"{column}.png")
        plt.savefig(save_path)
        plt.close()

    
    print(f"Graphs have been generated and saved in '{output_folder}' folder.")

if __name__ == "__main__":
    df_example = sns.load_dataset("penguins")

    auto_graphs(df_example)
```

2. Load the file matriu_distancies.xlsx into pandas, so that the row names and column names are those of the cities. Delete "Las Palmas de Gran Canaria" and "Palma" so that we can make the trip by car.

We are interested in visiting all the main cities in Spain covering the minimum possible            distance.
You don't have to do it optimally, we are interested in developing a reasonable solution using the tools you currently have.
For example, a simple (but not optimal) approach would be to always go to the nearest city that we haven't visited yet.
Make a function that, given the distance matrix and the city of origin, makes a route proposal that is as short as possible, returning a list with the order of visits. Also give the total distance traveled.
(EXTRA) From which city would the route be shortest with the proposed algorithm


```python
import pandas as pd
df1 = pd.read_excel(r"C:\Users\esrak\Downloads\Sprint10\matriu_distancies.xlsx",index_col=0)
df1
```

```python
df1 = df1.drop(index=["Las Palmas de Gran Canaria", "Palma"], errors='ignore')
df1 = df1.drop(columns=["Las Palmas de Gran Canaria", "Palma"], errors='ignore')
df1
```

```python
def nearest_neighbor_route(distance_matrix, start_city):
    visited = [start_city]
    total_distance = 0.0
    current_city = start_city

    # While there are cities left to visit
    while len(visited) < len(distance_matrix):
        distances = distance_matrix.loc[current_city]
        remaining = distances.drop(visited)
        next_city = remaining.idxmin()
        dist_to_next = remaining.min()

        total_distance += dist_to_next
        visited.append(next_city)
        current_city = next_city

    # Return to the starting city (optional, if we want a round trip)
    total_distance += distance_matrix.loc[current_city, start_city]
    visited.append(start_city)

    return visited, total_distance

route, total_dist = nearest_neighbor_route(df1, "Valencia")

print("Route:", " -> ".join(route))
print(f"Total distance: {total_dist:.2f} km")

results = {}

for city in df1.index:
    r, d = nearest_neighbor_route(df1, city)
    results[city] = d

best_start = min(results, key=results.get)

print(f"The shortest total route starts in {best_start} ({results[best_start]:.2f} km).")
```
