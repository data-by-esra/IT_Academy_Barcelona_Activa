1. Body Mass Index Calculator

```python
def bmi_calculator(weight, height):
    if weight <= 0 or height <= 0 or height >= 3:
        print("Please enter realistic, positive values for weight and height.")
        return

    bmi = weight / (height ** 2)
    print("Your BMI is:", round(bmi, 2))

    if bmi < 18.5:
        print("Underweight")
    elif bmi < 25:
        print("Normal weight")
    elif bmi < 30:
        print("Overweight")
    else:
        print("Obese")


def main():
    try:
        weight = float(input("Enter your weight (kg): "))
        height = float(input("Enter your height (in meters): "))
        bmi_calculator(weight, height)
    except ValueError:
        print("Invalid input. Please enter numeric values.")


if __name__ == "__main__":
    main()
```

2. Temperature Converter

```python
def convert_temperature(temp, unit):
    
    if not isinstance(temp, (int, float)):
        raise ValueError("Temperature must be a number.")
    if unit not in ["C", "F", "K"]:
        raise ValueError("Unit must be C, F, K")

    #Convert input to Celsius
    to_celsius = {
        "C": lambda x: x,
        "F": lambda x: (x - 32) * 5/9,
        "K": lambda x: x - 273.15
    }
    celsius_value = to_celsius[unit](temp)

    #Convert Celsius to other units
    from_celsius = {
        "C": lambda x: x,
        "F": lambda x: x * 9/5 + 32,
        "K": lambda x: x + 273.15
    }
    results = {unit: round(func(celsius_value), 2) for unit, func in from_celsius.items()}

    return results


def main():
    try:
        temp_input = input("Enter the temperature value: ").strip()
        if not temp_input.replace('.', '', 1).isdigit():
            raise ValueError("Please enter a valid number.")
        value = float(temp_input)

        unit = input("Enter the unit (C, F, K): ").strip().capitalize()
        results = convert_temperature(value, unit)

        print(results)
        return results

    except ValueError as e:
        print(f"Error: {e}")



conversions = main()
```

3. Word Counter of a Text

```python
import re

def analyze_text(text):
    #convert to lowercase and extract words(ignore puntuation)
    words = re.findall(r'\b\w+\b',text.lower())
    
    if not words:
        return "No words found in the text."
        
    
    #Count word occurrances
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word,0) + 1

    print(f"\nYou entered:\n{text}\n")    

    #Calculate average word lenght
    avg_lenght = sum(len(word) for word in words) / len(words) 

    print("Word Frequencies: ")
    for word,count in word_count.items():
        print(f"{word}:{count}")

    print(f"\nAverage word lenght: {avg_lenght:.2f}")

user_text = input("Enter a text: ")
analyze_text(user_text)
```

4. Reverse Dictionary (with possibility of duplicates)

```python
def reverse_dictionary(original_dict):
    """
    Reverses a dictionary by swapping keys and values.
    If duplicates exist, stores all associated keys in a list.
    """
    reversed_dict = {}

    for key, value in original_dict.items():
        if value in reversed_dict:
            if not isinstance(reversed_dict[value], list):
                reversed_dict[value] = [reversed_dict[value]]
            reversed_dict[value].append(key)
        else:
            reversed_dict[value] = key

    # Check for duplicates
    duplicates = {v: k for v, k in reversed_dict.items() if isinstance(k, list)}
    if duplicates:
        print("Warning: Duplicate values detected!")
        for value, keys in duplicates.items():
            print(f"Value '{value}' is associated with keys: {keys}")

    return reversed_dict



users = {"Alice Smith": 101, "Bob Johnson": 102, "Charlie Brown": 103, "Diana Prince": 104,
         "Eve Adams": 105, "Frank Miller": 105, "Grace Lee": 106, "Hannah Davis": 107,
         "Ian Clark": 108, "Jane Doe": 101}


reversed_users = reverse_dictionary(users)
print("\nReversed dictionary with duplicates preserved:")
print(reversed_users)
```

1. Word counter and sorter of a text

```python

from collections import Counter, defaultdict
import string


def get_words(text):
    translator = str.maketrans(string.punctuation + "¡¿",' ' * (len(string.punctuation)+ 2))
    text = text.translate(translator).lower()
    return text.split()

def word_counter_sorted_dict(filename):
    # Initialize full dictionary A-Z
    sorted_output = {letter: {} for letter in string.ascii_uppercase}


    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()

    words = get_words(text)

    #Count all words
    word_counts = Counter(words)

    # Group by first letter and sort
    temp_counter = defaultdict(dict)
    for word, count in word_counts.items():
        first_letter = word[0].upper()
        temp_counter[first_letter][word] = count

    # Sort each letter's words and fill the final dictionary
    for letter in temp_counter:
        sorted_output[letter] = dict(sorted(temp_counter[letter].items())
        )

    return sorted_output


filename = r"C:\Users\esrak\Downloads\Sprint9\tu me quieres blanca.txt"
result = word_counter_sorted_dict(filename)


for letter, words in result.items():
    print(f"{letter}: {words}")
```

2. Data type conversion

```python
data = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

def flatten(sequence):
    """Flatten a nested structure (lists, tuples)."""
    for item in sequence:
        if isinstance(item, (list, tuple)):
            yield from flatten(item)
        else:
            yield item

def split_by_convertibility(data):
    can_convert = []
    cannot_convert = []

    for item in flatten(data):
        try:
            can_convert.append(float(item))
        except (ValueError, TypeError):
            cannot_convert.append(item)

    return can_convert, cannot_convert

convertible, non_convertible = split_by_convertibility(data)

print("Convertible: ", convertible)
print("Not Convertible:", non_convertible)
```

1. Password Generator

```python
import random
import string
import pyperclip


def create_password(length, uppercase = True, lowercase = True, numbers = True, symbols = False, copy_to_clipboard = True ):

    char_pool = ''
    if uppercase:
        char_pool += string.ascii_uppercase
    if lowercase:
        char_pool += string.ascii_lowercase
    if numbers:
        char_pool += string.digits
    if symbols:
        char_pool += string.punctuation

    if not char_pool:
        raise ValueError("At least one character type must be selected.")
    

    password = []
    if uppercase:
        password.append(random.choice(string.ascii_uppercase))
    if lowercase:
        password.append(random.choice(string.ascii_lowercase))
    if numbers:
        password.append(random.choice(string.digits))
    if symbols:
        password.append(random.choice(string.punctuation))

    while len(password) < length:
        password.append(random.choice(char_pool))
    
    random.shuffle(password)
    password_str = ''.join(password[:length])

    if copy_to_clipboard:
        pyperclip.copy(password_str)
        print("Password copied to clipboard!")

        

    return password_str

if __name__ == "__main__":
    pwd = create_password(10, True, True, True, True)
    print("Generated password:", pwd)
    
```

2. Simple data processing

```python
from collections import defaultdict
import pandas as pd

filename = r"C:\\Users\\esrak\\Downloads\\Sprint9\\historic partits.txt"


stats = defaultdict(lambda: {"Wins": 0, "Draws": 0, "Losses": 0, "Goals For": 0, "Goals Against": 0, "Points": 0})

with open(filename, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        team1, result, team2 = line.split("\t")
        g1, g2 = map(int, result.split("-"))

        
        stats[team1]["Goals For"] += g1
        stats[team1]["Goals Against"] += g2
        stats[team2]["Goals For"] += g2
        stats[team2]["Goals Against"] += g1

       
        if g1 > g2:
            stats[team1]["Wins"] += 1
            stats[team2]["Losses"] += 1
            stats[team1]["Points"] += 3
        elif g1 < g2:
            stats[team2]["Wins"] += 1
            stats[team1]["Losses"] += 1
            stats[team2]["Points"] += 3
        else:
            stats[team1]["Draws"] += 1
            stats[team2]["Draws"] += 1
            stats[team1]["Points"] += 1
            stats[team2]["Points"] += 1


df = pd.DataFrame.from_dict(stats, orient="index").reset_index().rename(columns={"index": "Team"})


df = df[["Team", "Wins", "Draws", "Losses", "Goals For", "Goals Against", "Points"]]


df = df.sort_values(by=["Points", "Goals For"], ascending=[False, False]).reset_index(drop=True)


print("\n FINAL TABLE \n")
print(df.to_string(index=False))


top_scorer = df.loc[df["Goals For"].idxmax()]
most_conceded = df.loc[df["Goals Against"].idxmax()]

print("\n Team with most goals scored:")
print(f"{top_scorer['Team']} ({top_scorer['Goals For']} goals)")

print("\n Team that conceded the most goals:")
print(f"{most_conceded['Team']} ({most_conceded['Goals Against']} goals)")
```
